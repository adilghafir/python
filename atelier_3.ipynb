Définir une classe Segment dont le constructeur possède quatre paramètres : deux
pour l’origine et deux pour l’extrémité. Ce constructeur définit deux
attributs : orig et extrem, instances de la classe Point. De cette manière, vous concevez
une classe composite : la classe Segment est composée de deux instances de la

```python
class Vecteur2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vecteur2D({self.x}, {self.y})"

    def __add__(self, other):
        return Vecteur2D(self.x + other.x, self.y + other.y)

class Rectangle:
    def __init__(self, longueur=0, largeur=0):
        self.longueur = longueur
        self.largeur = largeur
        self.nom = "rectangle"

    def __str__(self):
        return f"{self.nom} de longueur {self.longueur} et de largeur {self.largeur}"

    def surface(self):
        return self.longueur * self.largeur

class Carre(Rectangle):
    def __init__(self, cote=0):
        super().__init__(cote, cote)
        self.nom = "carré"

class Point:
    def __init__(self, x=0.0, y=0.0):
        self.x = x
        self.y = y

class Segment:
    def __init__(self, x1=0.0, y1=0.0, x2=0.0, y2=0.0):
        self.orig = Point(x1, y1)
        self.extrem = Point(x2, y2)

    def __str__(self):
        return f"Segment de coordonnées ({self.orig.x}, {self.orig.y}) et ({self.extrem.x}, {self.extrem.y})"

# Programme principal
vecteur1 = Vecteur2D()
vecteur2 = Vecteur2D(1, 2)
print(vecteur1)
print(vecteur2)

vecteur3 = vecteur1 + vecteur2
print(vecteur3)

rectangle = Rectangle(15, 10)
print(rectangle)
print(rectangle.surface())

carre = Carre(5)
print(carre)
print(carre.surface())

segment = Segment(1, 2, 3, 4)
print(segment)
Vecteur2D(0, 0)
Vecteur2D(1, 2)
Vecteur2D(1, 2)
```
